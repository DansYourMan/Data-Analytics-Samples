```python
#Problem 1
import pandas

df_prob1 = pandas.read_csv('/MetObjects_Subset.csv')
df_prob1
```

```python
#Problem 2
df_prob2 = df_prob1.copy(deep='True')

# Step 1: Identify NULL columns
null_columns = df_prob2.columns[df_prob2.isnull().all()]

# Step 2: Remove NULL columns
df_prob2 = df_prob2.drop(columns=null_columns)

# Display the cleaned DataFrame
df_prob2
```

```python
#Problem 3
#From testing a strict equality will not pass NULL or NaN values
#Does 'Is Public Domain' equal True, ignore everything else
df_prob3 = df_prob2[df_prob2['Is Public Domain'] == True]

df_prob3
```

```python
#Problem 4
#Remove NULL values so the second stage works
df_filterNullRoles = df_prob3.dropna(subset=['Artist Role'])

#Searches the whole string for the 'Maker' substring
df_prob4 = df_filterNullRoles[df_filterNullRoles['Artist Role'].str.contains('Maker')]

df_prob4
```

```python
#Problem 5
import numpy as np

df_prob5 = df_prob4.copy(deep='True')

#Take the data from two other columns and do math
df_prob5['DecadesToMake'] = ((df_prob5['Object End Date'] - df_prob5['Object Begin Date']) / 10).apply(np.floor)

#Used for testing. I verified the math works as expected
#selected_columns = ['Object Begin Date', 'Object End Date', 'DecadesToMake']
#print(df_prob5.loc[:, selected_columns])

df_prob5
```

```python
#Problem 6
import matplotlib.pyplot as plt

# Get the top ten most common classifications
top_classifications = df_prob5['Classification'].value_counts().head(10)

# Plot the bar chart
top_classifications.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top Ten Most Common Classifications')
plt.xlabel('Classification')
plt.ylabel('Count')
plt.show()
```
