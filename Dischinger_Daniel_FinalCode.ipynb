```python
#Part 1
import pandas as pd

# Load the JSON file
df0 = pd.read_json('sub_class0.json')
df1 = pd.read_json('sub_class1.json')
df2 = pd.read_json('sub_class2.json')
df3 = pd.read_json('sub_class3.json')

# Discard rows with missing characteristics
df0.dropna(subset=['Title', 'id', 'Score', 'Num_Comments', 'Date'], inplace=True)
df1.dropna(subset=['Title', 'id', 'Score', 'Num_Comments', 'Date'], inplace=True)
df2.dropna(subset=['Title', 'id', 'Score', 'Num_Comments', 'Date'], inplace=True)
df3.dropna(subset=['Title', 'id', 'Score', 'Num_Comments', 'Date'], inplace=True)

# Discard the 'id' characteristic
df0.drop(columns=['id'], inplace=True)
df1.drop(columns=['id'], inplace=True)
df2.drop(columns=['id'], inplace=True)
df3.drop(columns=['id'], inplace=True)

# Save the modified DataFrame to a new JSON file
df0.to_json('sub_class0_P1.json')
df1.to_json('sub_class1_P1.json')
df2.to_json('sub_class2_P1.json')
df3.to_json('sub_class3_P1.json')

df0
```

```python
#Part 2
from textblob import TextBlob

# Calculate sentiment polarity of each title
df0['Sentiment'] = df0['Title'].apply(lambda x: TextBlob(x).sentiment.polarity)
df1['Sentiment'] = df1['Title'].apply(lambda x: TextBlob(x).sentiment.polarity)
df2['Sentiment'] = df2['Title'].apply(lambda x: TextBlob(x).sentiment.polarity)
df3['Sentiment'] = df3['Title'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Remove rows with neutral sentiment (-0.1 to 0.1 inclusive)
df0 = df0[(df0['Sentiment'] < -0.1) | (df0['Sentiment'] > 0.1)]
df1 = df1[(df1['Sentiment'] < -0.1) | (df1['Sentiment'] > 0.1)]
df2 = df2[(df2['Sentiment'] < -0.1) | (df2['Sentiment'] > 0.1)]
df3 = df3[(df3['Sentiment'] < -0.1) | (df3['Sentiment'] > 0.1)]

# Save the modified DataFrame to a new JSON file
df0.to_json('sub_class0_P2.json')
df1.to_json('sub_class1_P2.json')
df2.to_json('sub_class2_P2.json')
df3.to_json('sub_class3_P2.json')

df0
```

```python
#Part 3
import matplotlib.pyplot as plt

counts = []

counts.append(len(df0[df0['Sentiment'] > 0]))
counts.append(len(df0[df0['Sentiment'] < 0]))

counts.append(len(df1[df1['Sentiment'] > 0]))
counts.append(len(df1[df1['Sentiment'] < 0]))

counts.append(len(df2[df2['Sentiment'] > 0]))
counts.append(len(df2[df2['Sentiment'] < 0]))

counts.append(len(df3[df3['Sentiment'] > 0]))
counts.append(len(df3[df3['Sentiment'] < 0]))

# Create labels for the bars
labels = ['Class0 Positive', 'Class0 Negative',
          'Class1 Positive', 'Class1 Negative',
          'Class2 Positive', 'Class2 Negative',
          'Class3 Positive', 'Class3 Negative']

# Plot the bar chart
plt.figure(figsize=(20, 6))
plt.bar(labels, counts, color=['green', 'red'] * 4)
plt.title('Sentiment Counts by Class')
plt.xlabel('Class and Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
#Part 4
import pandas as pd

df0['Class'] = 0
df1['Class'] = 1
df2['Class'] = 2
df3['Class'] = 3

df_combined = pd.concat((df0, df1, df2, df3), ignore_index=True)

# Save the combined DataFrame to a JSON file
df_combined.to_json('sub_combined.json')
df_combined
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import confusion_matrix, accuracy_score

# Construct the feature array
feature_array = np.zeros((len(df_combined), 5))
feature_array[:, 0] = df_combined['Title'].str.len()  # Character length of the title
feature_array[:, 1] = df_combined['Sentiment']  # Sentiment of the title
feature_array[:, 2] = df_combined['Score']  # Score
feature_array[:, 3] = df_combined['Num_Comments']  # Number of comments
feature_array[:, 4] = df_combined['Date'].apply(lambda x: pd.to_datetime(x).year)  # Year from Date

#print(feature_array)

# Construct the target array
target_array = df_combined['Class'].values  # Class column as target

#print(len(target_array))

# Define the classifier (Logistic Regression in this example)
classifier = LogisticRegression(max_iter=1000, random_state=42)

# Perform 10-fold cross-validation and calculate accuracy
accuracy_scores = cross_val_score(classifier, feature_array, target_array, cv=10, scoring='accuracy')

# Calculate confusion matrix using cross_val_predict
predictions = cross_val_predict(classifier, feature_array, target_array, cv=10)
conf_matrix = confusion_matrix(target_array, predictions)

# Print accuracy scores and confusion matrix
print("10-Fold Cross-Validation Accuracy Scores:", accuracy_scores)
print("\nConfusion Matrix:")
print(conf_matrix)

# Calculate and print the mean accuracy
mean_accuracy = np.mean(accuracy_scores)
print(f"\nMean Accuracy: {mean_accuracy:.2f}")
```
